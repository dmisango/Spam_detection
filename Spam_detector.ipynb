```python
#Import the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer#turn text into feature values
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load the data
raw_data = pd.read_csv('/content/spam.csv')
raw_data
```

```python
raw_data.info()
```

```python
#label encoding for spam(0) and ham(1)
copy_data = raw_data.copy()

copy_data.loc[copy_data.Category == "spam","Category"] = 0 #spam will become 0
copy_data.loc[copy_data.Category == "ham","Category"] = 1 #ham will become 1

copy_data
```

```python
#Split your data into the test and train sets
X_train,X_test,Y_train,Y_test = train_test_split(copy_data.Message,copy_data.Category,test_size=0.25, random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.info())
print(Y_test.info())
```

```python
#feature extraction
feat_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
#min_df:if there is one occurrence it will not count(ignore); stop_words:ignore common english words repeated multiple times e.g, is, was; lowercase:convert all the letters to lowercase

#convert the messages
X_train_f = feat_extraction.fit_transform(X_train)
X_test_f = feat_extraction.transform(X_test)
```

```python
#convert the values of Y_train and Y_test to integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
print(Y_train.info())
print(Y_test.info())
```

```python
model =LogisticRegression()
```

```python
#training the model
model.fit(X_train_f,Y_train)
```

```python
#prediction test on the trained data
model.score(X_train_f,Y_train)
```

```python
#prediction test on testing data
pred1 = model.predict(X_test_f)
accuracy_score(Y_test,pred1)
```

```python
#Predicting mail
input_mail = ["As per your request 'Melle Melle (Oru Minnaminunginte Nurungu Vettam)' has been set as your callertune for all Callers. Press *9 to copy your friends Callertune"]

#convert to features
input_mail_f = feat_extraction.transform(input_mail)

#predict mail type
pred2 = model.predict(input_mail_f)
if pred2==1:
    print('mail is not spam')
else:
    print('mail is spam')
```
